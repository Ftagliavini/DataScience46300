```python
# Axes-level

sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
plt.show()

# explicando cada parte:

# sns.histplot(): Esto llama a la función histplot de Seaborn, que se utiliza para crear un histograma.

# data=penguins: Estás proporcionando los datos que queres graficar. 

# x="flipper_length_mm": Estás proporcionando la columna que queres utilizar como eje x.

# hue="species": Estás proporcionando la columna que queres utilizar como color, al tener 3 especies, seran 3 colores.

# "multiple="stack": Esta parte del código indica que queres apilar los histogramas de las diferentes especies en lugar de superponerlos."
```

```python
# Figure-level


sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
plt.show()

# explicando cada parte:

# sns.displot(): Esto llama a la función displot de Seaborn, que se utiliza para crear gráficos de distribución de datos, como histogramas o gráficos de densidad.

# data=penguins: Estás proporcionando los datos que queres graficar. 

# x="flipper_length_mm": Estás proporcionando la columna que queres utilizar como eje x.

# hue="species": Estás proporcionando la columna que queres utilizar como color.

# "multiple="stack": Esta parte del código indica que queres apilar los histogramas de las diferentes especies en lugar de superponerlos."
```

```python
# Multiples figuras (Figure-level)

plt.figure(figsize=(10, 4))# Al igual que en matplotlib, se puede cambiar el tamaño de la figura.
sns.displot(data=penguins, x="flipper_length_mm", hue="species", col="species")
plt.show()

# explicando cada parte:

# data=penguins: Estás proporcionando los datos que queres graficar. 

# x="flipper_length_mm": Estás proporcionando la columna que queres utilizar como eje x.

# hue="species": Estás proporcionando la columna que queres utilizar como color.

# col="species": Con esta opción, estás indicando que deseas crear múltiples gráficos, uno para cada especie de pingüino. 
```

 la principal diferencia radica en la versatilidad y el tipo de gráfico que puedes crear con cada función. 
 *  displot es más general y se utiliza para una variedad de gráficos de distribución.
 *  histplot se usa específicamente para crear histogramas. La elección entre ambas depende de tus necesidades y del tipo de gráfico que desees generar.

## FacetGrid 

Es una clase en Seaborn que te permite crear múltiples gráficos (subgráficos) basados en una o más variables categóricas en tu conjunto de datos. Estos subgráficos comparten el mismo eje y se organizan en una cuadrícula, lo que facilita la comparación de datos a través de diferentes categorías.

```python
# FacetGrid
tips = sns.load_dataset("tips")
sns.set(rc={'figure.figsize':(20, 8)}) 

# Creas un objeto FacetGrid llamado g basado en el DataFrame tips.
g = sns.FacetGrid(tips, col="time",  row="sex", height=4)  # Ajusta el valor de "height" según tus preferencias

# Especificas que deseas crear una cuadrícula de subgráficos con filas para la variable "sex" y columnas para la variable "time".

# utilizas g.map() para mapear un tipo de gráfico (en este caso, un gráfico de dispersión sns.scatterplot) en cada uno de los subgráficos de la cuadrícula.

g.map(sns.scatterplot, "total_bill", "tip")
```

El resultado es una cuadrícula de gráficos de dispersión que muestra cómo se relacionan "total_bill" y "tip" para diferentes combinaciones de "sex" y "time". Cada celda de la cuadrícula contiene un gráfico de dispersión que representa los datos para una combinación específica de "sex" y "time".

```python
g = sns.FacetGrid(tips, col="time", hue="sex", height=6)
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.add_legend()
```

```python
# Ejemplo aplicado
df_ejercicio = sns.load_dataset('exercise')
df_ejercicio = df_ejercicio.drop('Unnamed: 0', axis='columns')
df_ejercicio.head()
```

```python
# Filtrando donde "time" sea de 30 min.

df_30_min = df_ejercicio[df_ejercicio['time'] == '30 min']
df_30_min.head()
```

```python
# Figure -level

ax = sns.displot(data=df_30_min, x='pulse', kind='kde', hue='kind', fill=True)
ax.set(xlabel='Frecuencia Cardíaca', ylabel='Densidad', title='Distribución  de las pulsaciones')
plt.figure()


# NOTA: aparecieron cosas nuevas no? ok son ls siguientes:

# kind='kde' para crear un gráfico de densidad de kernel.

# hue='kind' para dividir los datos en categorías basadas en la columna "kind" y colorearlas.

# fill=True para rellenar el área bajo la curva en el gráfico de densidad.
```

```python
ax = sns.catplot(data=df_ejercicio, kind='violin', x='time', y='pulse', hue='diet', split=True)
ax.set(xlabel='Duración de ejercicio', ylabel='Frecuencia cardíaca', title='Categorización de la distribución de pulsaciones')
```
