## Metodos de correlacion

Cuando hacemos correlacion nos encontramos a priori con 3 metodos faciles de poder calcular

Estos metodos son:



*   [Pearson](https://es.wikipedia.org/wiki/Coeficiente_de_correlaci%C3%B3n_de_Pearson#:~:text=En%20estad%C3%ADstica%2C%20el%20coeficiente%20de,de%20medida%20de%20las%20variables) - usado para cuando sabemos su distribucion normal, ademas sensible a valores anomalos, muy usados con datasets muy grandes. NO TIENE OUTLIERS, Y TIENE QUE SER NORMAL
*   [Spearman](https://es.wikipedia.org/wiki/Coeficiente_de_correlaci%C3%B3n_de_Spearman) - Es para cualqueir tipo de variable dado que no necesita saberse sus distribucion, ademas muy usado para relacion de variables ordinales. Para datasets grandes. OUTLIERS  Y CUALQ DISTRIB

* [Kendall](https://statologos.com/kendalls-tau/) - Idem spearman pero para conjuntos chicos de datos DATASETS CHICOS = QUE SPEARMAN



Estos 3 metodos son usados en ciertas circunstancias la cual les explico arriba son los mas usados, lo importanta aqui es poder determinar en que momento usarse. Aunque el coeficiente de Pearson a priori es de lkos mas comunues , muchas veces en la realidad el de spearman se asemeja mas







```python
matrix_correlation = taxis.corr() #corr pearson por default
matrix_correlation
```

```python
#generacion de los graficos de correlacion entre los 3 metodos,
plt.figure(figsize=(36,12))
for j,i in enumerate(['pearson','kendall','spearman']):
  plt.subplot(1,3,j+1)
  correlation = taxis.dropna().corr(method=i)
  sns.heatmap(correlation, linewidth = 2)
  plt.title(i, fontsize=18)
  plt.yticks(fontsize=14)
  plt.xticks(fontsize=14)
```

como veran entre el metodo Spearman y Pearson hay valores diferentes , eso se da porque los valores de pearson son diferenciados por la distribucion que asumimos que es normal,  y seguramente no lo sea.
Para eso es importante ya en el analisis univariado sabersu distribucion


```python
#metodo para solo graficar en base a un threshold

mat_corr_spearman = taxis.corr('spearman')

high_corr = mat_corr_spearman[mat_corr_spearman>=0.8]
plt.figure(figsize=(12,8))
sns.heatmap(high_corr, cmap="Blues")
plt.title('Altas correlaciones');
```

```python
mat_corr_spearman = taxis.corr('spearman')

#filtro de la matrix
high_corr = mat_corr_spearman[mat_corr_spearman>=.8]

plt.figure(figsize=(12,8))
sns.heatmap(high_corr, cmap="Blues")
plt.title('Altas correlaciones');
```

```python
def corrFilter(x: pd.DataFrame, thres: float):
    #generate corr
    xCorr = x.corr('spearman')
    #filter corr by thres
    xFiltered = xCorr[((xCorr >= thres) | (xCorr <= -thres)) & (xCorr !=1.000)]
    #change dataframe format
    xFlattened = xFiltered.unstack().drop_duplicates().reset_index().sort_values(0, ascending= False).dropna()
    #rename columns
    xFlattened.columns = ['Variable_1', 'Variable_2', 'corr_value']
    return xFlattened

corrFilter(taxis, 0.7)

#x = taxis.corr('spearman')

#y = x[((x >= 0.7) | (x <= -0.7)) & (x !=1.000)]
#y.unstack().drop_duplicates().reset_index().sort_values(0).dropna()

#y.unstack().drop_duplicates().reset_index().sort_values(0, ascending= False).dropna()
```
