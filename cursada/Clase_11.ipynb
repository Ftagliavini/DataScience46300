# Carga de archivos

Reutilizaremos el dataframe que habiamos utilizado en la clase 9 con los precios de acciones del sector energetico

```python
print("Definiendo los simbolos de stock")
symbol_data_to_load = ['D','EXC','NEE','SO','DUK']
list_of_df = []

# Ciclo sobre simbolos
#llenar la lista de dataframes
print(" --- Inicio de Loop --- ")
for i in symbol_data_to_load:
    print("Procesando Simbolo: " + i)
    temp_df = pd.read_csv(i+'.csv',sep=',') # Cargamos los archivos CSV iterativamente
    temp_df['Volume_Millions'] = temp_df['Volume'] / 1000000.0 # Convertimos las unidades en millones
    temp_df['Symbol'] = i # Agregar nueva columna con el simbolo
    list_of_df.append(temp_df) # Cada data frame temporal es agregado a "list_of_df"

print(" --- Completado loop simbolos --- ")

# Combinar en un Dataframe unico usando concat todos los dataframe de la lista "list_of_df"
#permite pegar los dataframes de la lista
print("Agregando la data")
agg_df = pd.concat(list_of_df, axis=0)
agg_df
# Agregar estadisticas de retorno y volatilidad
# es mas rápido agregarlo al dataframe que a cada uno de los registros
print('Calculando estadisticas finales')

# VolStat calcula la volatilidad dividiendo la diferencia entre los precios más alto (High) y más bajo (Low) entre el precio de apertura (Open).
agg_df['VolStat'] = (agg_df['High'] - agg_df['Low']) / agg_df['Open']

# Return calcula el retorno diario como la relación entre el precio de cierre (Close) y el precio de apertura (Open), restando 1.
agg_df['Return'] = (agg_df['Close'] / agg_df['Open']) - 1.0

# Revisando la dimensión del dataframe
print("agg_df DataFrame dimension (filas, columnas): ")
print(agg_df.shape)

print("Head del DataFrame agg_df: ")
agg_df.head()
```

# Z score

```python
# Cargar datos
datos= agg_df[agg_df['Symbol']=='D']
l=[x for x in agg_df.columns if x not in ['Volume','Symbol']]
datos=datos[l]
datos['Date']=pd.to_datetime(datos['Date'])
datos.head()
```

```python
# Convertir a z score
datos_z=pd.DataFrame()
for j in datos.columns[1:]:
  # Dentro del bucle, se realiza el cálculo de los z-scores para cada columna.
  datos_z[j] = (datos[j] - datos[j].mean()) / datos[j].std()
datos_z['Fecha']= datos.Date
datos_z.head()
```

```python
datos_z['Open_x']= datos_z['Open'].apply(lambda x: 'Atipico' if (x>2 or x<-2) else 'Normal')
datos_z.head()
# aplica una función lambda a cada valor en la columna 'Open'.
# La función lambda toma un valor x, y si x es mayor que 2 o menor que -2, asigna el valor 'Atipico' a 'Open_x', de lo contrario, asigna 'Normal'.
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=datos_z.Fecha, y= datos_z.Open, hue= datos_z['Open_x'],s= 5)
```

# Distancia multiple

```python
# Extraer columnas de interes
data_multiple=datos_z.drop(columns=['Open_x','Fecha'])
data_multiple.head()
```

```python
def outlier_euclideano_d(x):
  # x: dataframe con valores numericos normalizados
  result_ = pd.Series([0] * len(x.iloc[:,1])) # Inicializamos una serie con ceros. (Aquí agregaremos los valores atípicos)
  data_mean = x.mean() # media de los datos
  dist = np.sqrt(np.sum(((x-data_mean) ** 2),axis=1)) #Distancia euclideana
  dist_mean = dist.mean() #media de las distancia
  dist_zscore = np.abs((dist - dist_mean) / dist.std())#z-score para las distancias

  # Se marca como 1 los valores cuyo z-score de distancia euclidiana sea mayor que 3,
  # lo que indica que están a más de 3 desviaciones estándar de la media de las distancias.
  # Estos valores se consideran atípicos.
  result_[((dist_zscore > 3))] = 1
  return result_
```

```python
# Crear una copia de los datos
euc_d = data_multiple.copy()
euc_d.head()
```

```python
# Aplicar la funcion
euc_d['outlier']=outlier_euclideano_d(euc_d)
euc_d.head(100)
```

```python
plt.figure(figsize=(15,14))
plt.subplot(2,3,1)
sns.scatterplot(x="Open",y="High",data=euc_d,hue="outlier",palette=["green","red"])
plt.subplot(2,3,2)
sns.scatterplot(x="Open",y="Low",data=euc_d,hue="outlier",palette=["green","red"])
plt.subplot(2,3,3)
sns.scatterplot(x="Open",y="Close",data=euc_d,hue="outlier",palette=["green","red"])
plt.subplot(2,3,4)
sns.scatterplot(x="Open",y="Adj Close",data=euc_d,hue="outlier",palette=["green","red"])
plt.subplot(2,3,5)
sns.scatterplot(x="Open",y="Volume_Millions",data=euc_d,hue="outlier",palette=["green","red"])
plt.subplot(2,3,6)
sns.scatterplot(x="Open",y="VolStat",data=euc_d,hue="outlier",palette=["green","red"])
```

```python
def parejas(df):
    '''Obtiene diagonal inferior de parejas en matriz de correlacion'''
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def correlaciones(data, n=5):
    au_corr = data.corr().abs().unstack()
    labels_to_drop = parejas(data)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]

print("Correlaciones mas grandes")
print(correlaciones(df, 10)) # Saquemos el top 10
```
