# Modelos de Decision Tree

Decision Tree es un algoritmo que se basa en arboles , los cuales tiene un sentido de pensamiento de if - else

La verdad como todo algoritmo se necesita que todos los datos sea numericos y ademas es un algortimo supervisado,  es decir que depende de una variable target, sea numerica o categorica



##explicacion

El arbol de decision es un estructura similar a una diagrama de flujo,

![decision tree](https://sitiobigdata.com/wp-content/uploads/2019/09/1569525641_839_Conceptos-b%C3%A1sicos-de-aprendizaje-autom%C3%A1tico-%C3%81rbol-de-decisi%C3%B3n-desde-cero-Parte-I.png)

Tenemos como cracteristicas:

* Nodo raíz (nodo de decisión superior ): Representa a toda la población o muestra y esto se divide en dos o más conjuntos homogéneos.

* División: Es un proceso de división de un nodo en dos o más subnodos.  Nodo de decisión: Cuando un subnodo se divide en subnodos adicionales, se llama nodo de decisión.

* Nodo de hoja / terminal: Los nodos sin hijos (sin división adicional) se llaman Hoja o nodo terminal.

* Poda: Cuando reducimos el tamaño de los árboles de decisión eliminando nodos (opuesto a la división), el proceso se llama poda.

* Rama / Subárbol: Una subsección del árbol de decisión se denomina rama o subárbol.

* Nodo padre e hijo: Un nodo, que se divide en subnodos se denomina nodo principal de subnodos, mientras que los subnodos son hijos de un nodo principal.




## Metodo de seleccion de division

Comunmente tenemos dos metodos de seleccion de cual es el atributo mejor para generar el nodo de decision

* Gini index
* Information Gain

![metodo de seleccion](https://sitiobigdata.com/wp-content/uploads/2019/09/1569525641_702_Conceptos-b%C3%A1sicos-de-aprendizaje-autom%C3%A1tico-%C3%81rbol-de-decisi%C3%B3n-desde-cero-Parte-I.png)



### Gini index

Es una funcion estadistica para calcular la homogeneidad de una funcion.

Mientras mayor es el numero de gini , mas homogeneo y mejor para la division del nodo.

Pasos:



1.   Calculo de gini para cada subnodo en base a (p<sup>2</sup> + q<sup> 2 </sup>)
2.   Ponderacion en base a indice de Gini de cada nodo en base a la division






### ejemplo gini

![ejemplo gini](https://sitiobigdata.com/wp-content/uploads/2019/09/1569525642_286_Conceptos-b%C3%A1sicos-de-aprendizaje-autom%C3%A1tico-%C3%81rbol-de-decisi%C3%B3n-desde-cero-Parte-I.png)


Division por genero

1.   Calculo de gini para subnodo Hembra = (0.2) * (0.2) + (0.8)*(0.8)
2.   Idem para subnodo Macho = (0.65) * (0.65) + (0.35)*(0.35)
3.   Gini ponderado = (10/30) * 0.68 + (20/30) * 0.55 = ***0.59***

___

Division por class

1.  Gini subnodo Clase IX = (0.43) *(0.43) + (0.57)*(0.57) = 0.51
2.  Gini subnodo Clase X = (0.56)*(0.56) + (0.44)*(0.44) = 0.51
3. Gini ponderado = (14/30) *0.51 + (16/30)*0.51 = ***0.51***



Por lo tanto la seleccion mas adecuada es de Genero



## Information Gain

Medida estadistica que caclula los niveles de impureza entre nodo padre y sus dos subnodos hijos , donde haya mayor diferencia mejor

nivel de impureza = entropia

Formula entropia = -p log2 p - q log2 q

formula Informatio Gain = entropia nodo principal - [AVG(entropia subnodos)]

### ejemplo Information Gain

Mismo ejemplo visual anterior

Entropia de nodo principal

-(0.5)log2(0.5) - (0.5)log2(0.5)

Division por genero

* Calculo de gini para subnodo Hembra = -(0.2) log2 (0.2) - (0.8) log2 (0.8) = 0.72.  
* Idem para subnodo Macho = - (0.65) log2 (0.65) - (0.35) log2 (0.35) = 0.93.  
* Gini ponderado = (10/30) * 0.72 + (20/30) * 0.93 = 0.86

Gain Information genero = 1 - 0.86 = ***0.14***

---

Division por class

* Gini subnodo Clase IX = -(0.43)log2(0.43) - (0.57)log2(0.57) = 0.99
* Gini subnodo Clase X = -(0.56)log2(0.56) - (0.44)log2(0.44) = 0.99
* Gini ponderado = (14/30) 0.99 + (16/30)0.99 = 0.99

Information Gain class = 1 - 0.99 = ***0.01***

El atributo a elegir es Genero por tener un Information Gain mayor

## Hiper parametros

Son parametros usados para regularizar el algortimo y proteger de sus deficiencias

Los mas usados:

* **max_depth**: la profundidad máxima del árbol
* **min_samples_split**: número mínimo de muestras necesarias antes de dividir este nodo. También se puede expresar en porcentaje.
* **min_samples_leaf**: número mínimo de muestras que debe haber en un nodo final (hoja). También se puede expresar en porcentaje.
* **max_leaf_nodes**: número máximo de nodos finales

## Virtudes

* Los árboles de decisión son fáciles de interpretar y visualizar.
* Puede capturar fácilmente patrones no lineales.
* Requiere menos preprocesamiento de datos por parte del usuario, por ejemplo, no es necesario normalizar las columnas.
* Se puede utilizar para ingeniería de características, como la predicción de valores perdidos, adecuada para la selección de variables.
* El árbol de decisión no tiene suposiciones sobre la distribución debido a la naturaleza no paramétrica del algoritmo.

#Deficiencias

* Datos sensibles al ruido, puede sobredimensionar los datos ruidosos.
* La pequeña variación en los datos puede dar lugar a un árbol de decisión diferente.
* Están sesgados con un conjunto de datos de desequilibrio, por lo que se recomienda equilibrar el conjunto de datos antes de crear el árbol de decisión
