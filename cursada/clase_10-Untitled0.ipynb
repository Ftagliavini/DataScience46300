```python
import pandas as pd
```

```python
data = {
    'Columna1': [1,2,3,4,5],
    'Columna2': [6,7,8,9,10]
}
print(data)
```

```python
df = pd.DataFrame(data)
df
```

```python
df['Suma'] = df['Columna1'] + df['Columna2']
# df.Suma = df.Columna1 + df.Columna2
df
```

```python
sumatoria = df.groupby("Suma").sum()
print(sumatoria)
```
