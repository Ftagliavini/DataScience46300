```python
df_lluvias = pd.read_csv('pune_1965_to_2002.csv')
df_lluvias.head()
```

```python
df_lluvias.index = df_lluvias['Year']
df_lluvias = df_lluvias.drop('Year', axis='columns')
df_lluvias.head()
```

```python
x = df_lluvias.index
y = df_lluvias.sum(axis='columns')
y
```

```python
fig, ax = plt.subplots(figsize=(12, 4)) #es el tamaño de la figura

ax.plot(x, y, label='Precipitaciones acumuladas') #dibuja la linea
ax.set_xlabel('Año') #es el nombre del eje x
ax.set_ylabel('Precipitación acumulada (mm.)') # es el nombre del eje y
ax.set_title('Precipitaciones acumuladas según el año') #es el titulo de la figura
ax.legend()
```

```python
fig, ax = plt.subplots() #crea un objeto figura y un objeto ax
mapeo_colores = ax.scatter(df_lluvias['Aug'], df_lluvias['Sep'], c=df_lluvias.index) #el objeto ax es el que va a tener la funcion scatter
fig.colorbar(mapeo_colores) # agrego el colorbar
ax.set_title('Precipitaciones Agosto-Septiembre') #agrego el titulo
ax.set_xlabel('Precipitaciones en Agosto (mm.)') #agrego el eje x
ax.set_ylabel('Precipitaciones en Septiembre (mm.)') #agrego el eje y
```

```python
fig, ax = plt.subplots(figsize=(8,4)) # figsize = tamaño de la figura
precipitaciones_acumuladas = df_lluvias.sum() # suma las precipitaciones por mes
ax.bar(df_lluvias.columns, precipitaciones_acumuladas) # grafica las barras
ax.set_title('Precipitaciones acumuladas desde 1965 a 2002, según el mes') # titulo
ax.set_ylabel('Precipitación total (mm.)') # etiqueta de eje y
ax.set_xlabel('Mes') # etiqueta de eje x
```
