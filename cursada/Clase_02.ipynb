##  TIPOS DE VARIABLES

nuestras variables pueden tomar diferentes tipos, las primitivas seran:
*   "integer" (entero)
*   "float" (flotante o decimal)
*   "string" (cadena de caracteres, escritura)
*   "Booleano" (verdadero o falso).

```python
# para valores "integer"

entero = 2
entero2 = 34

print(entero)
print(entero2)
```

```python
print(type(entero))
print(type(entero2))
```

```python
# para valores "float"

decimal = 2.000000000000
decimal2 = 3.1415

print(decimal)
print(decimal2)
```

```python
print(type(decimal))
print(type(decimal2))
```

```python
# para valores "string"

palabra = "mi cantidad de alumnos es"
palabra2 = "95"

print(palabra)
print(palabra2)
```

```python
print(type(palabra))
print(type(palabra2))
```

```python
# para valores "booleano"

booleano = True
booleano2 = False

print(booleano)
print(booleano2)
```

```python
print(type(booleano))
print(type(booleano2))
```
